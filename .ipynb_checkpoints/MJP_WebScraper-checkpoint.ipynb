# Web Scraper

This web scraper works specifically for the <a href="https://repository.library.brown.edu/studio/collections/id_592/">Modernist Journals Project in the Brown University digital repository</a>. While aspects of this might be useful as a model for other web scraping, there are a few details that are tailored just for the Brown repository. In particular, the for loops are designed with this repository in mind.

This code also saves the files onto my computer rather than streaming their contents directly into files. Like many other scrapers online, this one includes a slight delay to prevent congestion on the server side.

```python
from bs4 import BeautifulSoup
import re, time, requests, urllib.request
```

```python
# Functions

# Download MODS file: innermost layer.
def download_mods(recordPage):
    # Retrieve, name, and download record.
    recordName = "WebDirectory_Output/Mods/" + re.match(r'.*/bdr:(\d+)/.*', recordPage).group(1) + "-mods.xml"
    urllib.request.urlretrieve(recordPage, recordName)

# Download TEI file: innermost layer.
def download_record(recordPage):
    # Retrieve, name, and download record.
    recordName = "WebDirectory_Output/Text/" + re.match(r'.*/bdr:(\d+)/.*', recordPage).group(1) + ".xml"
    urllib.request.urlretrieve(recordPage, recordName)

# Access each record on page: middle layer.
def retrieve_record_links(page_url):
    response = requests.get(page_url)
    pageSoup = BeautifulSoup(response.text, "html.parser")
    link_list = []
    # Retrieve links to records from page.
    for record_link in pageSoup.select('a[href]'):
        regex = re.search(r'<a href="(/studio/item/bdr:\d+/)">', str(record_link))
        if regex is not None:
            link_list.append(regex.group(1))
        else:
            pass
    # Return unique values.
    return set(link_list)
```

```python
%%time

url = 'https://repository.library.brown.edu/studio/collections/id_592/?page='

# Open each page of MJP Repository in Brown Digital Repository.
for i in range(1, 100):
    page_url = url + str(i)
    record_links_list = retrieve_record_links(page_url)
    print ("Returning List from Page " + str(i))

    for link in record_links_list:
        recordPage = "https://repository.library.brown.edu" + link

        try:
            download_mods(recordPage + "MODS")
            download_record(recordPage + "TEI")
        except:
            print ("Could not find " + recordPage)
            pass

        time.sleep(1)
```
